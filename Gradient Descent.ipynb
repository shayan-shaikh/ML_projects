# Example one - a simple cost functions
## $f(x) = x^2 + x + 1$
 

```python
def f(x):
    return x**2 + x*2 + 1
```

```python
x_1 = np.linspace(-3,3,25)

```

```python
x_1
```

```python
plt.xlim(-3,3)
#plt.ylim(1,16)
plt.plot(x_1,f(x_1))
plt.xlabel("X values",fontsize = 16)
plt.ylabel("F(x)", fontsize = 16)
```

```python
plt.show()
```

## Derivative


```python
def der(x):
    return 2*x + 1
```

## Plotting derivative vs F(x)

```python
plt.figure(figsize=[5,12])
plt.subplot(2,1,1)

plt.plot(der(x_1), f(x_1), color='blue')
#plt.xlim(-4,6)
#plt.ylim(0,16)
plt.xlabel("derivative of x", fontsize=14)
plt.ylabel("f(x)", fontsize =14)

plt.subplot(2,1,2)
plt.plot((x_1), der(x_1), color='cyan', linewidth=3)
#plt.xlim(-4,6)
#plt.ylim(0,16)
plt.grid()
plt.xlabel("X", fontsize=14)
plt.ylabel("Derivative of X", fontsize =14)
plt.show()
```

## Grad descent!

```python
new_x = 3 #a new variable
prev_x = 0 #a temp previous value
step_mult = 0.1 #gradient descent factor / step multiplier
x_list = [new_x]
slope_list = [der(new_x)]
precision = 0.00001
for n in range(70):
    prev_x = new_x #equate em as we go right to left
    gradient = der(prev_x) #pass the temp value and store in gradient
    new_x = prev_x - step_mult*gradient
    x_list.append(new_x)
    slope_list.append(der(new_x))
    stepSize = abs(new_x - prev_x)
    if(stepSize < precision):
        print('Ended at', n)
        break
print('Local Minimum at', new_x)
print('slope :', der(new_x))
print('Cost :', f(new_x))
```

```python
x_list
```

```python
#plt.figure(figsize=[5,12])
plt.plot((x_1), f(x_1), color='blue')
#plt.xlim(-4,6)
#plt.ylim(0,16)
plt.xlabel("x", fontsize=14)
plt.ylabel("f(x)", fontsize =14)
v = np.array(x_list)
plt.scatter(x_list,f(v),color='red')
plt.show()
```

```python
plt.plot(der(x_1),f(der(x_1)), color='cyan', linewidth=3)
#plt.xlim(-4,6)
#plt.ylim(0,16)
plt.grid()
plt.xlabel("Derivative", fontsize=14)
plt.ylabel("Function of Derivative", fontsize =14)
sv = np.array(slope_list)
plt.scatter(sv,f(sv),color='red',alpha=0.3,s=100)
plt.show() 
```

```python
plt.plot((x_1), der(x_1), color='cyan', linewidth=3)
#plt.xlim(-4,6)
#plt.ylim(0,16)
plt.grid()
plt.xlabel("X", fontsize=14)
plt.ylabel("Derivative of X", fontsize =14)
plt.scatter(v,der(v), color='blue',s=50,alpha=0.5)
plt.show()
```

## Example 2: Multiple Minima vs Initial Guess 
$g(x) = x^4 - 4x^2 + 5$

```python
x_2 = np.linspace(-2,2,1000)
```

```python
def g(x):
    return x**4 - 4*x**2 + 5
```

```python
def derg(x):
    return 4*x**3 - 8*x
```

```python
plt.figure(figsize=[13,5])
plt.subplot(1,2,1)

plt.plot((x_2), g(x_2), color='blue')
plt.xlim(-2,2)
plt.ylim(0.5,6)
plt.xlabel("x", fontsize=14)
plt.ylabel("g(x)", fontsize =14)

plt.subplot(1,2,2)
plt.plot((x_2), derg(x_2), color='cyan', linewidth=3)
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.grid()
plt.xlabel("X", fontsize=14)
plt.ylabel("Derivative of X", fontsize =14)
plt.show()
```

### Gradient descent with Python Fn

```python
def grad_desc(derFunc, initguess, multiplier=0.02, precision=0.001):
    new_x = initguess
    x_list = [new_x]
    slope_list = [derFunc(new_x)]
    step_mult = multiplier
    for n in range(70):
        prev_x = new_x 
        gradient = derFunc(prev_x)
        new_x = prev_x - step_mult*gradient
        x_list.append(new_x)
        slope_list.append(derFunc(new_x))
        stepSize = abs(new_x - prev_x)
        if(stepSize < precision):
            break
    return new_x, x_list, slope_list
```

```python
local_min,list_x,der_list = grad_desc(derFunc = derg,initguess = 1)
print(local_min)
print('total steps :', len(list_x))
```

```python
plt.plot(list_x,der_list)
```
